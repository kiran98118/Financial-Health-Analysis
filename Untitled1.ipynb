```python
import pandas as pd

df = pd.read_csv("/Users/saikiran/Downloads/financial_health_analysis_dataset.csv")  

df.head()
df.info()
df.describe()
```

```python
#Scenario 1: Testing if the average monthly profit is greater than $10,000
#Hypotheses: Null Hypothesis (H₀): The average monthly profit ≤ $10,000, Alternative Hypothesis (H₁): The average monthly profit > $10,000


from scipy import stats


benchmark = 10000
t_stat, p_value = stats.ttest_1samp(df['Profit'], benchmark)

print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

# Interpretation
alpha = 0.05  # 5% significance level
if p_value < alpha:
    print("Reject H₀: The average monthly profit is significantly greater than $10,000.")
else:
    print("Fail to reject H₀: Not enough evidence to claim profit > $10,000.")

    
#Conclusion: Since p-value < 0.05, we reject H₀. So, The average monthly profit is significantly greater than $10,000.
```

```python
# Scenario 2: Comparing Revenue and Expenses using a Paired t-test.
#Hypotheses: Null Hypothesis (H₀): There is no significant difference between revenue and expenses. Alternative Hypothesis (H₁): Revenue and expenses are significantly different.


t_stat, p_value = stats.ttest_rel(df['Revenue'], df['Expenses'])

print(f"\nScenario 2 Results:")
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject H0: Revenue and expenses are significantly different.")
else:
    print("Fail to reject H0: No significant difference between revenue and expenses.")
    
#Conclusion: Since p-value < 0.05, we reject H₀. So, Revenue and expenses are significantly different.
```

```python
#Scenario 3: Yearly Revenue and Profit Analysis

df['year'] = pd.to_datetime(df['Date']).dt.year  

# Calculate Yearly Revenue and Profit
yearly_data = df.groupby('year')[['Revenue', 'Profit']].sum()

# Calculate Year-over-Year Growth
yearly_data['Revenue Growth'] = yearly_data['Revenue'].pct_change() * 100
yearly_data['Profit Growth'] = yearly_data['Profit'].pct_change() * 100

print(yearly_data)
```

```python
Summary of key findings:
    *Profit is significantly greater than $10,000 (T-test result, p < 0.05).
    *Revenue and expenses are significantly different, meaning expenses do not fully consume revenue.
    *Year-over-Year revenue and profit growth can be calculated after correcting the column name error.    
```
